A loop is used to repeat a block of code multiple times until a condition is met.

        
                Types of loop
                1. For loop
                2. While loop
    
    For loop in python
  .A for loop in Python is used to iterate over a sequence (such as a list, tuple, string, or range) and execute a block of code for each element in that sequence.

```python
for i in range(5):
    print(i)
```

```python
for i in range(2,22,2):
    print(i)
```

```python
student = {
    "name": "Ayush",
    "class": "second year",
    "branch": "computer science",
    "roll_no": 12345
}
```

```python
student
```

```python
for x in student:
    print(x)
```

```python
for items in student.items():
    print(items)
```

```python
for x in range(10):
    if x==5:
        break   
    print(x)
```

```python
for x in range(10):
    if x==5:
        continue  
    print(x)
```

### while loop
A while loop is used to repeat a block of code as long as a given condition is true.

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```
